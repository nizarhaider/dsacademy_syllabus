### 1. Integer (int)
Integers are whole numbers, positive or negative, without decimals.

````python
age = 35

print(type(age))
````

### 2. Float
Floats are numbers that contain a decimal point.



````python
# Example of a float
height = 5.28638273

print(height)
print(type(height))
````

### 3. String (str)
Strings are sequences of characters, enclosed in quotes.


````python
# Example of a string
name = "I love Sri Lanka"

print(name)
print(type(name))
````

### 4. Boolean (bool)
Booleans represent one of two values: True or False.

````python
# Example of a boolean
is_student = False

print(is_student)
print(type(is_student))
````

### 5. Tuple
Tuples are ordered collections of items, which can be of different types. They are enclosed in parentheses and are immutable.

````python
# Example of a tuple
favorite_colors = ("blue", "green", "red")

print(favorite_colors)
print(type(favorite_colors))
````

### 6. List
Lists are ordered collections of items, which can be of different types. They are enclosed in square brackets.

````python
# Example of a list
grades = [88, 92, 75, 91]

favourite_movies = ["Godzilla","Behind Her Eyes","Titanic"]


print(grades[0])
print(type(grades[0]))
````

### 7. Dictionary (dict)
Dictionaries are unordered collections of key-value pairs. They are enclosed in curly braces.

````python
# Example of a dictionary
personal_info = {
    "Name": "Nizar",
    "age": 26,
    "Country": "Sri Lanka"
}

print(personal_info)
````

# Learning Basic Arithmetic Operations in Python

In this notebook, we will learn about basic arithmetic operations in Python. These operations include addition, subtraction, multiplication, division, and more. Let's get started!

### 1. Addition (`+`)

Addition is used to add two or more numbers together.



````python
# Example of addition
a = 5
b = 3

sum_result = a + b
print(f"The sum of {a} and {b} is {sum_result}")
````

````python
a = 10
b = 12

print(a+b)
````

### 2. Subtraction (`-`)
Subtraction is used to subtract one number from another.

````python
# Example of subtraction
a = 10
b = 4
difference = a - b
print(f"The difference between {a} and {b} is {difference}")
````

### 3. Multiplication (`*`)
Multiplication is used to multiply two or more numbers.

````python
# Example of multiplication
a = 7
b = 6

product = a * b
print(f"The product of {a} and {b} is {product}")
````

````python
a = 5
b =10

print(a * b)
````

### 4. Division (`/`)
Division is used to divide one number by another. It returns a float.

````python
# Example of division
a = 15
b = 3
quotient = a / b

print(f"The quotient of {a} divided by {b} is {quotient}")
````

### 5. Modulus (`%`)
The modulus operator is used to find the remainder of the division of one number by another.

It is useful for determining whether a number is even or odd (remainder when divided by 2).


````python
numerator = 10
denominator = 3

remainder = numerator % denominator
print(f"The remainder when {numerator} is divided by {denominator} is: {remainder}")
````

````python
print(12 % 3)
````

## Learning Control Structures and Functions in Python

Control structures allow you to manage the flow of your program, while functions help you organize and reuse code.

#### 1. Conditional Statements (`if`, `elif`, `else`)

Conditional statements are used to perform different actions based on different conditions.




````python
# Example of if, elif, else
marks = 55

if (marks > 70):
  print("Pass")
elif (marks > 50 and marks < 70):
  print("Barely Passed")
else: print("Fail")
````

### 2. Loops


#### a. `for` Loop

The for loop is used to iterate over a sequence (such as a list, tuple, or string).

````python
# Example of for loop
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
````

````python
for num in range(10):
  print("Hello everyone")
````

#### b. `while` Loop

The while loop repeats as long as a specified condition is true.

````python
# Starts from 0 and ends in 4 which is a total of 5 iterations
x = 0

while x < 5:
  print(f"The current value of x is {x}")
  x += 1
````

### Functions
Functions are reusable pieces of code that perform a specific task. You can define your own functions to organize your code and make it more readable.

1. Defining and Calling Functions
- You define a function using the def keyword, followed by the function name and parameters.

````python
# Defining a function
def greet(name):
    print(f"Hello, {name}!")

# Calling the function
greet("Alice")
````

````python
# Your code here
def add_numbers(a, b):
    return a + b

# Test the function
result = add_numbers(5, 7)
print(result)  # Output: 12
````

````python
# Create function
def add_numbers(a,b):
  return a + b

# Call the function
a = 10
b = 10
output = add_numbers(a,b)
print(output)
````
