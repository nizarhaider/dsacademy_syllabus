
# EDA 101

## Introduction

Exploratory Data Analysis (EDA) is a critical step in the data science process. It involves understanding and summarizing the main characteristics of a dataset, often visualizing these properties. EDA can help uncover insights, identify patterns, establish relationships and trends, and inform feature engineering and modeling strategies.



## Loading Data

To start our EDA, we first need to load the dataset. We'll use the `pandas` library, a powerful tool for data manipulation and analysis in Python.


```python
import pandas as pd 

df_sales = pd.read_csv('sales_data.csv')
# Given dataset for sales
df_sales.head()
```


# Loading the dataset from the CSV file

```python

# Checking data types of each column
data_types = df_sales.dtypes

# Getting summary statistics for the dataset
summary_statistics = df_sales.describe()

# Checking for missing values
missing_values = df_sales.isnull().sum()

data_types, summary_statistics, missing_values
```


## Univariate Analysis

Univariate analysis involves examining one variable (or feature) at a time. This can help in understanding the distribution of the variable, spotting outliers, and identifying patterns.



### Histogram for `Customer_Age`

A histogram is a graphical representation that organizes a group of data points into user-specified ranges. It's useful to understand the distribution of a numerical variable.


```python

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df_sales['Customer_Age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.grid(axis='y')
plt.show()
```


### Pie Chart for `Payment_Method`

Pie charts help visualize the distribution of categorical data by dividing a circle into proportional segments.



### Box Plot for `Price`

A box plot (or whisker plot) displays a summary of a set of data values. It can help in visualizing the distribution of the data, spotting outliers, and comparing multiple data sets.


```python

plt.figure(figsize=(10, 6))
plt.boxplot(df_sales['Price'], vert=False)
plt.title('Box Plot for Price')
plt.xlabel('Price')
plt.grid(axis='x')
plt.show()
```

```python

payment_methods = df_sales['Payment_Method'].value_counts()

plt.figure(figsize=(10, 6))
payment_methods.plot.pie(autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightpink', 'lightyellow'])
plt.title('Distribution of Payment Methods')
plt.ylabel('')  # This is to remove the default 'Payment_Method' ylabel
plt.show()
```


## Bivariate and Multivariate Analysis

Bivariate analysis involves examining two variables to determine the empirical relationship between them. Multivariate analysis involves examining more than two variables.



### Scatter Plot between `Price` and `Units_Sold`

Scatter plots are great for visualizing the relationship between two numerical variables.


```python

plt.figure(figsize=(10, 6))
plt.scatter(df_sales['Price'], df_sales['Units_Sold'], color='lightcoral', alpha=0.5)
plt.title('Scatter Plot between Price and Units Sold')
plt.xlabel('Price')
plt.ylabel('Units Sold')
plt.grid(True)
plt.show()
```


## Handling Missing Data

Handling missing data is a crucial step in the EDA process. Different techniques can be applied depending on the nature of the data and the type of analysis being conducted.



### Correlation Heatmap

A heatmap of correlations can give insights into relationships between multiple variables in the dataset. A correlation value close to 1 implies a strong positive correlation, meaning as one variable increases, the other also tends to increase. A correlation close to -1 implies a strong negative relationship.


```python

import seaborn as sns

correlation_matrix = df_sales.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Correlation Heatmap')
plt.show()
```


### Handling Missing Data in `Customer_Age`

There are several strategies to handle missing data. Some of the common methods include:

1. **Deletion**: Remove rows with missing values.
2. **Imputation**: Fill missing values with a specific value, such as mean, median, mode, or a constant.
3. **Predictive Modeling**: Use machine learning algorithms to predict and fill missing values.

For this exercise, we'll use imputation to fill missing values in the `Customer_Age` column with its median.


```python

# Imputing missing values in 'Customer_Age' with median
median_age = df_sales_missing['Customer_Age'].median()
df_sales_missing['Customer_Age'].fillna(median_age, inplace=True)

# Checking for missing values after imputation
missing_values_after_imputation = df_sales_missing['Customer_Age'].isnull().sum()
missing_values_after_imputation
```


## Outlier Detection and Handling

Outliers are data points that differ significantly from other observations. They can arise due to variability in the data or errors. Detecting and handling outliers is crucial as they can significantly impact the results of the analysis.



Outliers can be identified in several ways:

1. **Statistical Methods**: Using measures like Z-scores or the IQR (Interquartile Range).
2. **Visualization Tools**: Box plots and scatter plots can visually highlight data points that stand out from the rest.

Once identified, you can decide to remove or adjust these outliers, or leave them if they're valid points.


## Feature Engineering

Feature engineering involves creating new features from the existing ones or transforming features to improve their usability in modeling. It can be essential for improving model performance.



### Creating a New Feature: `Revenue`

Let's create a new feature called `Revenue`, which is the product of `Price` and `Units_Sold`. This can give us insights into the total sales value for each transaction.


```python

df_sales['Revenue'] = df_sales['Price'] * df_sales['Units_Sold']
df_sales[['Price', 'Units_Sold', 'Revenue']].head()
```


## Data Transformations

Data transformations can be applied to make the data suitable for analysis or to fit assumptions of certain algorithms. Common transformations include normalization, standardization, and logarithmic transformations.



### Log Transformation of `Price`

Logarithmic transformation can be useful when dealing with skewed data. It can help in reducing the impact of outliers and make the distribution more normal-like.


```python

# Applying log transformation to 'Price' column
df_sales['Log_Price'] = np.log1p(df_sales['Price'])  # Using log1p to also handle zero values

# Plotting the original and transformed data
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Original Price distribution
ax[0].hist(df_sales['Price'], bins=30, color='skyblue', edgecolor='black')
ax[0].set_title('Original Price Distribution')
ax[0].set_xlabel('Price')
ax[0].set_ylabel('Number of Transactions')
ax[0].grid(axis='y')

# Log-transformed Price distribution
ax[1].hist(df_sales['Log_Price'], bins=30, color='salmon', edgecolor='black')
ax[1].set_title('Log-transformed Price Distribution')
ax[1].set_xlabel('Log-transformed Price')
ax[1].set_ylabel('Number of Transactions')
ax[1].grid(axis='y')

plt.tight_layout()
plt.show()
```
