
# Cloud 101

## Introduction

Cloud computing refers to the delivery of computing services—including servers, storage, databases, networking, software, analytics, and intelligence—over the Internet ("the cloud") to offer faster innovation, flexible resources, and economies of scale.

Historically, individuals and companies would run applications or programs from software downloaded on a physical computer or server in their building. Cloud computing allows people access to the same kinds of applications through the internet.




## Benefits of Cloud Computing

There are numerous advantages to adopting cloud computing:

- **Cost-Efficiency**: Eliminates the capital expense of buying hardware and software and setting up on-site datacenters.
- **Scalability and Flexibility**: Quickly scale up or down to match the demand without affecting performance.
- **Accessibility and Collaboration**: Access data and applications from anywhere, promoting collaboration.
- **Disaster Recovery and Business Continuity**: Cloud providers offer backup solutions ensuring data safety.




## Deployment Models

Based on the needs of users, several cloud computing deployment models have been developed:

- **Public Cloud**: Owned and operated by third-party cloud service providers, which deliver their computing resources over the Internet. 
- **Private Cloud**: Used exclusively by a single business or organization.
- **Hybrid Cloud**: Combines public and private clouds, allowing data and applications to be shared between them.
- **Community Cloud**: Shared by several organizations and supports a specific community that has shared concerns.




## Service Models

Cloud computing is categorized into different service models, each serving a unique purpose:

- **Infrastructure as a Service (IaaS)**: Provides virtualized computing resources over the internet.
- **Platform as a Service (PaaS)**: Provides a platform allowing customers to develop, run, and manage applications.
- **Software as a Service (SaaS)**: Provides software applications over the internet on a subscription basis.
- **Function as a Service (FaaS)**: Serverless computing, runs individual functions as they're needed without provisioning an application.




## Key Cloud Providers

Several companies offer cloud computing services:

- **Amazon Web Services (AWS)**: Offers a mix of IaaS, PaaS, and SaaS services.
- **Microsoft Azure**: Provides a set of cloud services that developers and IT professionals use to build, deploy, and manage applications.
- **Google Cloud Platform (GCP)**: Known for high-end computing solutions and robust machine learning tools.
- **Others**: Include IBM Cloud, Oracle Cloud, Alibaba Cloud, and more.




## Basic Cloud Operations

For those new to cloud services, here are some basic operations:

- **Setting up an account with a cloud provider**: This usually involves visiting the provider's website, signing up with an email and providing credit card details (for billing purposes).
- **Navigating the cloud dashboard**: Once logged in, users are presented with a dashboard that provides access to various services and resources.
- **Deploying a simple application or service**: Depending on the provider, users can quickly launch virtual machines, host web applications, or utilize databases.

*Note: This is an overview, and practical walkthroughs are recommended for hands-on experience.*




## Security in the Cloud

Security remains a top concern in cloud computing:

- **Shared Responsibility Model**: While cloud providers ensure the security of the cloud, customers are responsible for the security of what they put in the cloud.
- **Encryption**: Data should be encrypted both in transit and at rest.
- **Firewalls and Identity Management**: Proper firewalls should be set up, and identity and access management policies should be in place to ensure only authorized access.
