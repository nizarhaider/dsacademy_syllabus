# Python Data Types & Variables

## Numbers (integers, floats)

In Python, there are several types of numbers, but the most common ones are integers and floating-point numbers.


```python

# Integer
integer_value = 10
print(f"Integer value: {integer_value}")

# Float
float_value = 20.5
print(f"Float value: {float_value}")
```


## Strings

Strings in Python are used to represent textual data. They can be created using single, double, or triple quotes.


```python

# String examples
string_single_quotes = 'This is a string in single quotes.'
string_double_quotes = "This is a string in double quotes."

print(string_single_quotes)
print(string_double_quotes)
```


# Basic Operators

## Arithmetic operators

Arithmetic operators are used to perform mathematical operations.


```python

# Demonstration of arithmetic operators
x = 15
y = 4

# Output: x + y = 19
print(f"x + y = {x+y}")

# Output: x - y = 11
print(f"x - y = {x-y}")

# Output: x * y = 60
print(f"x * y = {x*y}")

# Output: x / y = 3.75
print(f"x / y = {x/y}")
```


# Control Structures

## Conditional statements (if, elif, else)

In Python, you can use `if`, `elif`, and `else` statements to run code based on a specific condition.


```python

# Example of using if, elif, and else
number = 5

if number > 0:
    result = "positive"
elif number < 0:
    result = "negative"
else:
    result = "zero"

result
```


## Loops (for, while)

Loops allow you to execute a block of code repeatedly.


```python

# Example of using a for loop to iterate over a list
fruits = ["apple", "banana", "cherry"]
fruit_output = []

for fruit in fruits:
    fruit_output.append(fruit)

fruit_output
```


## List comprehensions

List comprehensions provide a concise way to create lists.


```python

# Using list comprehension to create a list of squares from 0 to 9
squares = [x**2 for x in range(10)]
squares
```


# Functions

## Defining a function

In Python, the `def` keyword is used to declare functions.


```python

# Example of defining a function
def greet(name):
    return f"Hello, {name}!"

greeting = greet("Alice")
greeting
```


## Calling a function

Once a function is defined, it can be called by its name followed by parentheses.



## Return statement

The `return` statement is used to exit a function and return a value.


```python

# Example of a function with a return statement
def add_numbers(a, b):
    return a + b

sum_result = add_numbers(5, 3)
sum_result
```

# Basic Exception Handling

## try, except blocks

The `try` block allows you to test a block of code for errors, and the `except` block lets you handle the error.


```python

# Demonstrating the use of try and except
try:
    result = 10 / 0
except ZeroDivisionError:
    result = "Cannot divide by zero!"

result
```


## Handling specific exceptions

You can specify multiple `except` blocks to handle different exceptions separately.



## finally block

The `finally` block, if specified, will be executed regardless of whether an exception occurs or not.



## Lambda functions

Lambda functions are small anonymous functions. They can have any number of arguments, but can only have one expression. The expression is evaluated and returned when the lambda function is called.


```python

# Example of a lambda function to double a number
double = lambda x: x * 2

double_result = double(5)
double_result
```


## Using functions from a module

Once a module is imported, its functions can be accessed using the dot notation.


```python

# Using the factorial function from the math module
factorial_result = math.factorial(5)
factorial_result
```


## Writing to a file

You can write to a file either by creating a new one or by appending to an existing one.


```python

# Writing to a file (this will create a new file or overwrite an existing one)
file_path_write = "/mnt/data/sample_file.txt"
with open(file_path_write, "w") as file:
    file.write("Hello, Data Science Academy!")

file_path_write  # Display the path to the file we just wrote to
```


## Appending to a file

To add content to an existing file without overwriting the previous content, you can use the 'a' mode.


```python

# Appending to the file we just created
with open(file_path_write, "a") as file:
    file.write("
Appended text!")

# Reading the file to display its contents
with open(file_path_write, "r") as file:
    appended_content = file.read()

appended_content
```


# Popular Python Libraries for Data Science

Python's ecosystem is rich with libraries that are tailored for a wide range of data science tasks. Here are some of the most popular and widely used libraries:

1. **NumPy**: A foundational package for numerical computations in Python. It provides support for large multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays.

2. **pandas**: A powerful library for data manipulation and analysis. It provides data structures like DataFrames and Series for handling and analyzing structured data.

3. **Matplotlib**: A plotting library for creating static, animated, and interactive visualizations in Python.

4. **Seaborn**: Built on top of Matplotlib, it provides a higher-level interface for creating visually appealing statistical graphics.

5. **Scikit-learn**: A machine learning library that provides simple and efficient tools for data mining and data analysis.

6. **TensorFlow and PyTorch**: Libraries for deep learning that provide a set of flexible tools for building and training complex neural network models.

7. **Statsmodels**: Used for estimating and testing statistical models.

8. **SciPy**: Used for high-level computations. It builds on the capabilities of NumPy and provides additional modules for optimization, integration, and other scientific computing tasks.

9. **NLTK and spaCy**: Libraries for natural language processing tasks.

10. **Beautiful Soup and Scrapy**: For web scraping.

It's worth noting that while these libraries are powerful, the best tool often depends on the specific task and the needs of the user.



# Best Practices for Python in Data Science

When using Python for data science, adhering to certain best practices can help in writing efficient, readable, and maintainable code. Here are some recommended best practices:

1. **Start with Exploratory Data Analysis (EDA)**: Before diving into modeling, always explore your data. Visualizations, summary statistics, and data cleaning are essential steps.

2. **Use Version Control**: Tools like Git can help you track changes, collaborate with others, and revert to previous versions of your code when needed.

3. **Write Functions**: If you find yourself writing the same code more than once, consider wrapping it in a function. This not only reduces redundancy but also makes your code more readable.

4. **Comment and Document**: Write comments for complex code sections. Use docstrings to document functions, methods, and classes.

5. **Follow the PEP 8 Style Guide**: It's the official style guide for Python and provides conventions for writing code in Python.

6. **Use Virtual Environments**: Virtual environments allow you to maintain separate spaces for your projects, each with its own dependencies.

7. **Regularly Backup Your Code and Data**: Always ensure your data and code are backed up. Cloud solutions like GitHub, Bitbucket, or cloud storage services can be useful.

8. **Stay Updated**: The Python and data science ecosystem is always evolving. Stay updated with the latest versions and methodologies.

9. **Optimize When Necessary**: While it's tempting to optimize your code constantly, it's often better to write clear and straightforward code first, then optimize later if needed.

10. **Test Your Code**: Especially for complex projects, write tests to ensure your code works as expected. Libraries like `pytest` can be helpful.

Remember, while these best practices provide a guideline, the most important thing is to find a workflow that suits your needs and ensures productivity.
