
# APIs 101

## Introduction

An API, or Application Programming Interface, is a set of rules and protocols that allow different software applications to communicate with each other. 
Imagine it as a waiter taking your order (request) to the kitchen (server) and then bringing back your food (response).

APIs play a crucial role in modern software development, enabling the integration of diverse systems, data sharing, and the extension of functionalities.




## Types of APIs

There are several types of APIs, each serving a different purpose:

1. **Web APIs**: These are accessed over the internet using HTTP or HTTPS protocol. Examples include the Twitter API, Google Maps API, and more.
2. **Database APIs**: Allow communication between an application and a database. For instance, SQL is an API for DBMS.
3. **Operating System APIs**: These enable applications to request services from the operating system, such as reading a file or opening a network connection.




## Basic Terminologies

Before diving deeper, let's familiarize ourselves with some basic terminologies:

- **Endpoint**: The specific address (URL) where an API can be accessed.
- **Request & Response**: A request is made to fetch or send data to the API, and in return, the API sends back a response.
- **HTTP methods**: 
  - **GET**: Retrieve data.
  - **POST**: Send data.
  - **PUT**: Update data.
  - **DELETE**: Remove data.
- **Status codes**: Indicate the result of the HTTP request. For example, 200 means 'OK', and 404 means 'Not Found'.




## Practical Example: Using a Public API

Let's use a simple public API called JSONPlaceholder for demonstration. It's a free fake online REST API used for testing and prototyping.

### Making a GET request to fetch data


```python

import requests

# Making a GET request
response = requests.get("https://jsonplaceholder.typicode.com/posts")

# Checking if the request was successful
if response.status_code == 200:
    data = response.json()
    print(data[0])  # Print the first post
else:
    print("Failed to retrieve data.")
```


### Understanding the response

The response will be a list of posts in JSON format. Each post will have an ID, title, and content among other details.



## Best Practices and Common Pitfalls

When working with APIs, it's essential to be aware of best practices and potential pitfalls:

- **Rate limiting**: Many APIs have a limit to how many requests you can make in a given time frame. Always check an API's documentation for its rate limits.
- **API keys and authentication**: Some APIs require authentication using API keys or other methods. Always keep your API keys secret.


```python

# Intentionally making a request to a non-existent endpoint to demonstrate error handling
url = "https://jsonplaceholder.typicode.com/non_existent_endpoint"

response = requests.get(url)

if response.status_code == 200:
    print(response.json())
else:
    print(f"Error {response.status_code}: {response.text}")
```


### Handling errors

Ensure your code can handle errors gracefully. Check the response's status code and handle non-200 responses appropriately.
